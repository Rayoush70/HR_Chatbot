```python
sequences2=tokenizer2.texts_to_sequences(Y)
# pad sequences 
max_sequence_len = max([len(x) for x in sequences2])
output_sequences = np.array(pad_sequences(sequences2,maxlen=max_sequence_len,padding='post'))
output_sequences , output_sequences.shape
```

```python
# ONE_HOT ENCODING
import tensorflow as tf
ys = tf.keras.utils.to_categorical(output_sequences, num_classes=total_words2)
ys.shape 
```

```python
embedding_dim=128
from tensorflow.keras.layers import Embedding, Dense, RNN,GRU,InputLayer,TimeDistributed
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
model = Sequential([
    InputLayer(input_shape=(19,)),
    Embedding(total_words,128,embeddings_initializer=tf.keras.initializers.random_normal(),
              input_length=19,trainable=True,mask_zero=True),
    GRU(256,return_sequences=True),
    TimeDistributed(Dense(33,activation='softmax'))])
model.compile(loss='categorical_crossentropy',
              optimizer="Adam",    
              metrics=["accuracy"])
model.summary()
```
